# subplots()

```python
type(plt.subplots())
```

```python
fig, axes = plt.subplots()

axes.plot(x, y, 'b')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
```

```python
axes
```

```python
index = 1
for current_ax in axes:
    current_ax.plot(x, y, 'b')
    current_ax.set_title('Title ' + str(index))
    current_ax.set_xlabel('X')
    current_ax.set_ylabel('Y')
    index += 1

fig.tight_layout()
fig
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x, y, 'b')
axes[1].plot(y, x, 'g')

fig.tight_layout()
```

# Figuresize, Aspect Ratio, DPI

```python
# figsize is a tuple of the width and height in inches
# dpi is dots per inch ( pixels per inch )

fig = plt.figure(figsize=(8, 4), dpi = 100)
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
```

```python
fig = plt.figure(figsize=(8, 4), dpi = 50)
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, 
                        figsize=(12, 3), dpi = 100)

axes[0].plot(x, y, 'r')
axes[1].plot(y, x, 'g')

fig.tight_layout()
```

Saving Images


```python
fig.savefig("plot1.png", dpi = 100)
```
